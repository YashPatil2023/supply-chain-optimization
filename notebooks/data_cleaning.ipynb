# Data Cleaning & Exploratory Data Analysis (EDA)

This notebook demonstrates the exploratory data analysis and data cleaning steps performed on both the Walmart Sales Forecasting Dataset and the Retail Supply Chain Dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.style.use('ggplot')
```

## 1. Walmart Dataset

```python
# Load clean merged data
walmart_df = pd.read_csv('../data/walmart/walmart_merged.csv', parse_dates=['Date'])
print(walmart_df.shape)
walmart_df.head()
```

```python
# Visualize sales distribution
plt.figure(figsize=(10, 5))
sns.histplot(walmart_df['Weekly_Sales'], bins=100, kde=True)
plt.xlim(-5000, 100000)
plt.title('Distribution of Weekly Sales')
plt.xlabel('Weekly Sales ($)')
plt.show()
```

## 2. Retail Supply Chain Dataset

```python
# Load retail data
retail_df = pd.read_csv('../data/supply_chain/retail_clean.csv')
print(retail_df.shape)
retail_df.head()
```

```python
# Visualize Lead Time
plt.figure(figsize=(8, 4))
sns.countplot(data=retail_df, x='Lead_Time_Days', hue='Ship Mode')
plt.title('Lead Time Distribution by Ship Mode')
plt.show()
```
