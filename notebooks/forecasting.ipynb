# Demand Forecasting Modeling

This notebook trains and evaluates the Gradient Boosting machine learning model for demand forecasting.

```python
import pandas as pd
import matplotlib.pyplot as plt
import json

# Load evaluation metrics generated by the pipeline
with open('../outputs/model_metrics.json', 'r') as f:
    metrics = json.load(f)

print(f"Validation MAE:  ${metrics['val_mae']:.2f}")
print(f"Validation RMSE: ${metrics['val_rmse']:.2f}")
print(f"Validation R²:   {metrics['val_r2']:.4f}")
```

```python
# Feature Importance
fi = pd.Series(metrics['feature_importance'])
fi.sort_values(ascending=True).tail(10).plot(kind='barh', figsize=(8, 6), color='#3498db')
plt.title('Top 10 Feature Importances')
plt.xlabel('Importance')
plt.show()
```

## Validation Predictions vs Actuals

```python
val_df = pd.read_csv('../outputs/validation_predictions.csv', parse_dates=['Date'])
val_df = val_df.sort_values('Date')

# Let's check a specific store & dept
store_dept = val_df[(val_df['Store'] == 1) & (val_df['Dept'] == 1)]

plt.figure(figsize=(12, 5))
plt.plot(store_dept['Date'], store_dept['Weekly_Sales'], label='Actual', marker='o')
plt.plot(store_dept['Date'], store_dept['Predicted_Sales'], label='Predicted', marker='x', linestyle='--')
plt.title('Actual vs Predicted Weekly Sales (Store 1, Dept 1)')
plt.legend()
plt.show()
```
